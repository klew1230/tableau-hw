```python
import pandas as pd
import numpy as np
```

```python
_Spring201903 = "Resources/2019_Spring/201903-citibike-tripdata.csv"
_Spring201904 = "Resources/2019_Spring/201904-citibike-tripdata.csv"
_Spring201905 = "Resources/2019_Spring/201905-citibike-tripdata.csv"
```

```python
Spring201903_df = pd.read_csv(_Spring201903)
Spring201904_df = pd.read_csv(_Spring201904)
Spring201905_df = pd.read_csv(_Spring201905)
```

```python
Spring201903_df.insert(loc=0, column='date', value=201903)
```

```python
Spring201904_df.insert(loc=0, column='date', value=201904)
```

```python
Spring201905_df.insert(loc=0, column='date', value=201905)
```

```python
# append csvs

merge_Spring2019 = pd.concat([Spring201903_df, Spring201904_df, Spring201905_df])
Spring2019 = merge_Spring2019.drop(columns=['starttime', 'stoptime', "start station id", "end station id", "bikeid"])
Spring2019.head()
```

```python
Spring2019.dropna(inplace = True )
```

```python
Spring2019.dtypes
```

```python
Spring2019['gender'] = np.where(Spring2019['gender']==0,"Other",(np.where(Spring2019['gender']==1,"Male","Female")))

Spring2019.head()
```

```python
Spring2019['gender'] = Spring2019['gender'].apply(str)
Spring2019.dtypes
```

```python
Spring2019.head()
```

```python
Spring2019.to_csv('Resources/Spring2019.csv', index=False)
```
