```python
import pandas as pd
import numpy as np
```

```python
_Spring202003 = "Resources/2020_Spring/202003-citibike-tripdata.csv"
_Spring202004 = "Resources/2020_Spring/202004-citibike-tripdata.csv"
_Spring202005 = "Resources/2020_Spring/202005-citibike-tripdata.csv"
```

```python
Spring202003_df = pd.read_csv(_Spring202003)
Spring202004_df = pd.read_csv(_Spring202004)
Spring202005_df = pd.read_csv(_Spring202005)
```

```python
Spring202003_df.insert(loc=0, column='date', value=202003)
```

```python
Spring202004_df.insert(loc=0, column='date', value=202004)
```

```python
Spring202005_df.insert(loc=0, column='date', value=202005)
```

```python
# append csvs

merge_Spring2020 = pd.concat([Spring202003_df, Spring202004_df, Spring202005_df])
Spring2020 = merge_Spring2020.drop(columns=['starttime', 'stoptime', "start station id", "end station id", "bikeid"])
Spring2020.head()
```

```python
Spring2020.dropna(inplace = True )
```

```python
Spring2020['gender'] = np.where(Spring2020['gender']==0,"Other",(np.where(Spring2020['gender']==1,"Male","Female")))

Spring2020.head()
```

```python
Spring2020['gender'] = Spring2020['gender'].apply(str)
Spring2020.dtypes
```

```python
Spring2020.to_csv('Resources/Spring2020.csv', index=False)
```
