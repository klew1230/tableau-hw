```python
import pandas as pd
import numpy as np
```

```python
Spring2019 = "Resources/Spring2019.csv"
Spring2020 = "Resources/Spring2020.csv"
Spring2021 = "Resources/Spring2021.csv"
Spring2022 = "Resources/Spring2022.csv"
```

```python
Spring2019_df = pd.read_csv(Spring2019)
Spring2020_df = pd.read_csv(Spring2020)
Spring2021_df = pd.read_csv(Spring2021)
Spring2022_df = pd.read_csv(Spring2022)
```

```python
# append 19&20

merge_19_20 = pd.concat([Spring2019_df, Spring2020_df])
merge_19_20.head()
```

```python
# rename columns

merge_19_20 = merge_19_20[['date', 'tripduration', 'start station name', 'start station latitude', 'start station longitude', 'end station name', 'end station latitude', 'end station longitude', 'usertype', 'birth year', 'gender']]
merge_19_20.columns = ['date', 'tripduration', 'start station name', 'start station latitude', 'start station longitude', 'end station name', 'end station latitude', 'end station longitude', 'member_casual', 'birth year', 'gender']

merge_19_20.head()
```

```python
# rename data points
merge_19_20 = merge_19_20.replace(to_replace='Subscriber', value='member')
merge_19_20 = merge_19_20.replace(to_replace='Customer', value='casual')

merge_19_20.head()
```

```python
merge_19_20.to_csv('Spring_19thru20.csv', index=False)
```
